```python
import matplotlib.pyplot as plt
import seaborn as sns


def f(x):
    return x**2 - 2

def secant(x0, x1, tol=1e-6):
    while abs(x1 - x0) > tol:
        x2 = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))
        x0, x1 = x1, x2
    return x1

x0, x1 = 1, 2
x_vals = [x0, x1]

while abs(x1 - x0) > 1e-6:
    x2 = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))
    x0, x1 = x1, x2
    x_vals.append(x2)

sns.set(style="darkgrid")
plt.figure(figsize=(8, 4))
sns.lineplot(x=range(len(x_vals)), y=x_vals, marker='o', label='x values')
plt.axhline(y=x_vals[-1], color='red', linestyle='--', label=x_vals[-1])
plt.title("Secant Method Convergence")
plt.xlabel("Iteration")
plt.ylabel("x value")
plt.show()
```
