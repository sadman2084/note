```python
import math
import matplotlib.pyplot as plt
import numpy as np
def func(x):
    return  3*x -math.cos(x)-1

def bisection(a, b, tolerance):
    while abs(b-a)/b>=tolerance:
        c=(a+b)/2
        if(func(c)*func(a) == 0):
            return c
        elif func(c)*func(a) > 0:
            a = c
        else:
            b = c
    print("ans is ",c)  

a=1
b=2
tolerance=0.01
if a > b:
    print("a is greater than b")
else:
    bisection(a, b, tolerance)

x = np.linspace(a, b, 400)
y = [func(i) for i in x]
plt.axhline(0, color='black', linewidth=0.5) #for x axis
plt.axvline(0, color='black', linewidth=0.5) #for y axis
plt.plot(x, y, label='f(x) = 3x - cos(x) - 1', color='blue')
plt.title('Graph of the function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
