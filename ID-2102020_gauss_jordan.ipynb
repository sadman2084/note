```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

matrix = [
    [2, -1, 3, 9],
    [1, 0, -2, -3],
    [3, 2, -1, 4]
]

def print_matrix(matrix):
    for row in matrix:
        print(*row)

def gauss_jordan(matrix):
    n = len(matrix)

    for i in range(n):
        if matrix[i][i] == 0:
            for j in range(i + 1, n):
                if matrix[j][i] != 0:
                    matrix[i], matrix[j] = matrix[j], matrix[i]
                    break
            else:
                return 1 

        pivot = matrix[i][i]
        for k in range(n + 1):
            matrix[i][k] /= pivot

        for j in range(n):
            if i != j:
                factor = matrix[j][i]
                for k in range(n + 1):
                    matrix[j][k] -= factor * matrix[i][k]

    return 0 

def print_result(matrix):
    print("Solutions:")
    for row in matrix:
        print(row[-1], end=" ")
    print()

flag = gauss_jordan(matrix)

print("\nFinal Matrix:")
print_matrix(matrix)

if flag == 1:
    print("\nNo unique solution.")
else:
    print_result(matrix)


```

```python
solution = [row[-1] for row in matrix]

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(solution)+1), solution, marker='o', linestyle='-', color='blue')
plt.xticks(range(1, len(solution)+1))
plt.xlabel("Variable (x1, x2, x3)")
plt.ylabel("Value")
plt.title("Gauss-Jordan Solution Line Graph")
plt.grid(True)
plt.show()
```
