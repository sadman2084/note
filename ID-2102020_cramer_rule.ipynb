```python
import numpy as np

A = np.array([
    [2, 1, -1],
    [-3, -1, 2],
    [-2, 1, 2]
])
B = np.array([8, -11, -3])
det_A = np.linalg.det(A)

if det_A == 0:
    print("No unique solution exists!")
else:
    solutions = []
    for i in range(len(B)):
        A_copy = A.copy()
        A_copy[:, i] = B
        det_Ai = np.linalg.det(A_copy)   
        solutions.append(det_Ai / det_A)
    x, y, z = solutions
    print(f"x = {x}, y = {y}, z = {z}")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


A = np.array([
    [2, 1, -1],
    [-3, -1, 2],
    [-2, 1, 2]
])
B = np.array([8, -11, -3])

# det(A)
det_A = np.linalg.det(A)

if det_A == 0:
    print("No unique solution exists!")
else:
    solutions = []
    det_values = []

   
    plt.figure(figsize=(4,3))
    sns.heatmap(A, annot=True, cmap="coolwarm", cbar=True)
    plt.title("Coefficient Matrix A")
    plt.show()

    for i in range(len(B)):
        A_copy = A.copy()
        A_copy[:, i] = B
        det_Ai = np.linalg.det(A_copy)   
        solutions.append(det_Ai / det_A)
        det_values.append(det_Ai)


        plt.figure(figsize=(4,3))
        sns.heatmap(A_copy, annot=True, cmap="coolwarm", cbar=True)
        plt.title(f"Matrix A_{i+1} (Column {i+1} replaced with B)")
        plt.show()

    x, y, z = solutions
    print(f"x = {x}, y = {y}, z = {z}")


    plt.figure(figsize=(4,3))
    sns.barplot(x=["x", "y", "z"], y=[x, y, z], palette="viridis")
    plt.title("Solution Values")
    plt.ylabel("Value")
    plt.show()


    plt.figure(figsize=(4,3))
    sns.barplot(x=["det(A)", "det(A1)", "det(A2)", "det(A3)"],
                y=[det_A] + det_values, palette="mako")
    plt.title("Determinants Comparison")
    plt.ylabel("Value")
    plt.show()
```
