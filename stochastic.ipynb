```python
from sklearn.linear_model import SGDRegressor
import numpy as np


X = np.array([1, 2, 3]).reshape(-1, 1)  
y = np.array([2, 4, 6])


model = SGDRegressor(random_state=42,max_iter=1000,learning_rate='constant')


model.fit(X, y)


print("Learned weight (w):", model.coef_[0])
print("Predictions:", model.predict(X))
```
