```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
def func(x):
    return 3*x - math.cos(x) - 1

def falsi(a,b,tolerance):
    while (abs(b-a)/b) >=tolerance:
        c = b - (func(b) * (a - b)) / (func(a) - func(b))
        if func(a)*func(c)==0:
            return c
        elif  func(a)*func(c)>0:
            a = c
        else:
            b = c
    print("ans is ", c)
    return c


a,b,tolerance=map(float,input("enter a b and tolerance ").split())
c=falsi(a,b,tolerance)

xs=np.linspace(a,b,100)
ys=[func(x) for x in xs]

sns.set(style='darkgrid')
plt.axhline(0,color='black',lw=0.5)
sns.lineplot(x=xs,y=ys)

plt.scatter(c,func(c),color='red',s=50)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('False Position Method')
plt.show()
```
