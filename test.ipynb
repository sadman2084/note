```python
import math
import matplotlib.pyplot as plt

def g(mu, T1, T2, theta):
    return 0.02 * mu + (1/theta) * math.log(T2 / T1)

T1 = 50    
T2 = 100    
theta = 3.14 


mu = 0.3             
tolerance = 1e-6    
max_iterations = 50 


iterations = []
mu_values = []


print("Fixed Point Iteration Method for Pulley Friction")
print(f"{'Iter':<5}{'μ (current)':<15}{'μ (next)':<15}{'Error':<15}")

for i in range(1, max_iterations + 1):
    mu_next = g(mu, T1, T2, theta)
    error = abs(mu_next - mu)
    
    iterations.append(i)
    mu_values.append(mu)
    
    print(f"{i:<5}{mu:<15.6f}{mu_next:<15.6f}{error:<15.6f}")
    
    if error < tolerance:
        print("\nConverged successfully!")
        break
    mu = mu_next


print(f"\nEstimated coefficient of friction (μ) ≈ {mu_next:.6f}")


plt.figure(figsize=(8,5))
plt.plot(iterations, mu_values, marker='o', linestyle='-', linewidth=2)
plt.title('Convergence of μ using Fixed Point Iteration', fontsize=14)
plt.xlabel('Iteration Number', fontsize=12)
plt.ylabel('μ Value', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
