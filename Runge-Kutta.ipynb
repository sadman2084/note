```python
#example: dy/dx = x + y,  y(0) = 1

import numpy as np
import matplotlib.pyplot as plt


def f(x, y):
    return x + y


def runge_kutta(f, x0, y0, h, n):
    xs = [x0]
    ys = [y0]
    x, y = x0, y0

    for i in range(n):
        k1 = h * f(x, y)
        k2 = h * f(x + h/2, y + k1/2)
        k3 = h * f(x + h/2, y + k2/2)
        k4 = h * f(x + h, y + k3)

        y = y + (k1 + 2*k2 + 2*k3 + k4) / 6
        x = x + h

        xs.append(x)
        ys.append(y)

    return np.array(xs), np.array(ys)



x0 = 0
y0 = 1
h = 0.1       
n = 20         

xs, ys = runge_kutta(f, x0, y0, h, n)


for i in range(len(xs)):
    print(f"x = {xs[i]:.2f}, y = {ys[i]:.5f}")


plt.plot(xs, ys, "o-", label="RK4 Solution")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Runge-Kutta (Kutta) Method Example")
plt.grid(True)
plt.legend()
plt.show()
```
