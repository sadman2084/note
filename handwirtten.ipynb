```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
# ANN apply er age joto value ase tader range 0 and 1 er moddhe anbo
# karon value beshi difference thakle converge late hbe
# taile joto gula value ase take 255 diye divide krobo
# karon max 255 hole 255/255 mane 1 hbe.mane shob value 0 to 1 er modhe chole ashbe
x_train=x_train/255
x_test=x_test/255
```

```python
x_train[0]
```

```python
plt.imshow(x_train[0])
```

```python
# so ekhon ANN banabo
# but image gula ase 28*28 mane 784 te nai
# 784 te banate hole flatten use krbo jeta side by side niye ashbe shob pixel k
#mane 2D image k 1D image banabe

model=Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
 
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,epochs=10,validation_split=0.2)
```

```python
y_preb=model.predict(x_test)
```

```python
y_pred=y_preb.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
